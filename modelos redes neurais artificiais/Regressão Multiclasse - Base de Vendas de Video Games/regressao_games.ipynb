### Regressao multiclasse para preços de video games

```python
import pandas as pd
```

#### Carregamento da base de dados

```python
from keras.layers import Dense, Dropout, Activation, Input
from keras.models import Model
```

```python
base = pd.read_csv('../../data/video-games-sales/Video_Games_Sales_as_at_22_Dec_2016.csv')
```

### Preprocessamento da base de dados
- Remover ruidos e informações irrelevantes

```python
base
```

```python
base = base.drop('Other_Sales', axis=1)
base = base.drop('Global_Sales', axis=1)
base = base.drop('Developer', axis=1)
```

```python
base
```

### Tratamento de valores nulos
- deletar registro com valores nulos

```python
base = base.dropna(axis=0)
```

- deletar registros com vendas menor do que 1 na America e na Europa

```python
base = base.loc[base['NA_Sales']<1]
base = base.loc[base['EU_Sales']<1]
```

```python
base['Name'].value_counts()
```

```python
nome_jogos = base.Name
```

```python
base = base.drop('Name', axis=1)
```

```python
base
```

### Extrair Previsores e as classes

```python
previsores = base.iloc[:,[0,1,2,3,7,8,9,10,11]].values
```

```python
venda_na = base.iloc[:,4].values
venda_eu = base.iloc[:,5].values
venda_jp = base.iloc[:,6].values
```

```python
previsores[0]
```

- Transformação dos dados

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
```

```python
previsores[:,0] = label_encoder.fit_transform(previsores[:,0])
```

```python
previsores[:,2] = label_encoder.fit_transform(previsores[:,2])
previsores[:,3] = label_encoder.fit_transform(previsores[:,3])
previsores[:,8] = label_encoder.fit_transform(previsores[:,8])
```

```python
previsores[0]
```

 - Categorização

```python
from sklearn.compose import ColumnTransformer 
```

```python
# s 1 0
# r 0 1
ct = ColumnTransformer([("", OneHotEncoder(),[0,2,3,8])], remainder="passthrough") # The last arg ([0]) is the list of columns you want to transform in this step
previsores = ct.fit_transform(previsores)
```

```python
previsores
```

### Criação da rede neural 

```python
camada_entrada = Input(shape=(301,))
camada_oculta1 = Dense(units=152, activation = 'sigmoid')(camada_entrada)
camada_oculta2 = Dense(units=152, activation='sigmoid')(camada_oculta1)
camada_saida1 = Dense(units=1, activation = 'linear')(camada_oculta2)
camada_saida2 = Dense(units=1, activation = 'linear')(camada_oculta2)
camada_saida3 = Dense(units=1, activation='linear')(camada_oculta2)
```

```python
regressor = Model(inputs = camada_entrada, outputs = [camada_saida1, camada_saida2, camada_saida3])
```

```python
regressor.compile(optimizer = 'adam', loss='mse')
```

#### Treinamento do modelo

```python
regressor.fit(previsores, [venda_na, venda_eu, venda_jp], epochs = 100, batch_size = 100)
```

#### Previsão dos preços

```python
previsao_na, previsa_eu, previsao_jp = regressor.predict(previsores)
```
