### Validação Cruzada para a base plantas de iris

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

### Carregar a base

```python
base = pd.read_csv('iris.csv')
```

- extrair os elementos previsores e as classes

```python
previsores = base.iloc[:, 0:4].values
```

```python
classe = base.iloc[:,4].values
```

### Transformação dos dados
- Necessário transformar os elementos da classe em valores numéricos e categóricos

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

- Transformação para valor numérico

```python
classe= labelEncoder.fit_transform(classe)
```

- Transformação para valor categórico

```python
classe_dummy = np_utils.to_categorical(classe)
```

### Estrutura da rede neural artificial

```python
def criarRede():
    classificador = Sequential()
    classificador.add(Dense(units = 4, activation = 'relu', input_dim = 4))
    classificador.add(Dense(units = 4, activation = 'relu'))
    classificador.add(Dense(units = 3, activation = 'softmax'))
    classificador.compile(optimizer = 'adam', loss= 'categorical_crossentropy', metrics = ['categorical_accuracy'])
    return classificador
```

- Aplicacao da validação cruzada

```python
classificador = KerasClassifier(build_fn = criarRede, epochs = 1000, batch_size=10)
resultados = cross_val_score(estimator = classificador, X = previsores, y = classe, cv = 10, scoring = 'accuracy')
```

## Resultados
- média das acurácias
- desvio padrão das acurácias

```python
media = resultados.mean()
desvio = resultados.std()
```

```python
print(media, desvio)
```
