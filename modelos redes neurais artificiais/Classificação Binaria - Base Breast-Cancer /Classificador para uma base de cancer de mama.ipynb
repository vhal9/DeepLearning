### Classificador para uma base de cancer de mama utilizando redes neurais artificiais

#### Carregar base de dados como dataframe pandas
- previsores contém os dados sobre caracteristicas de cada item.
- classes contém a classificação de cada item (0 para negativo e 1 para positivo).

```python
import pandas as pd
```

```python
previsores= pd.read_csv('../../data/base breast cancer/entradas-breast.csv')
```

```python
classes = pd.read_csv('../../data/base breast cancer/saidas-breast.csv')
```

#### Divisão da base de dados em treinamento e teste
- 25% para teste

```python
from sklearn.model_selection import train_test_split
```

```python
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classes,test_size=0.25 )
```

```python
classe_treinamento[:8]
```

#### Treinamento do modelo utilizando Keras

```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
import numpy as np
```

- número de neurônios nas camadas ocultas.
- fórmula = (número de características + 1) / 2

```python
num_neuronios_camada_saida = np.around((len(previsores_treinamento.columns)+1)/2)
```

```python
num_neuronios_camada_saida
```

#### Definiçao da camada de entrada e construção das camadas ocultas
- função de ativação nas camadas ocultas = relu

```python
classificador = Sequential()
```

```python
classificador.add(Dense(units = int(num_neuronios_camada_saida), activation='relu', kernel_initializer 
                        = 'random_uniform', input_dim = 30))
```

```python
classificador.add(Dense(units = int(num_neuronios_camada_saida), activation='relu', kernel_initializer 
                        = 'random_uniform'))
```

```python
classificador.add(Dense(units = int(num_neuronios_camada_saida), activation='relu', kernel_initializer 
                        = 'random_uniform'))
```

#### Definição da camada de saída
- função de ativação na camada de saída = sigmoid

```python
classificador.add(Dense(units=1, activation='sigmoid'))
```

```python
otimizador = keras.optimizers.Adam(lr = 0.00001, decay = 0.00001, clipvalue=0.5)
```

```python
classificador.compile(optimizer=otimizador, loss = 'binary_crossentropy', metrics=['binary_accuracy'])
```

```python
#classificador.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['binary_accuracy'])
```

```python
classificador.fit(previsores_treinamento, classe_treinamento, batch_size=10, epochs=800, verbose=0)
```

```python
previsao = classificador.predict(previsores_teste)
```

```python
previsao = (previsao > 0.5)
```

```python
previsao
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
precisao = accuracy_score(classe_teste, previsao)
```

```python
precisao
```

```python
pd.value_counts(classe_teste.values.flatten())
```

```python
matriz = confusion_matrix(classe_teste, previsao)
```

```python
resultado = classificador.evaluate(previsores_teste, classe_teste)
```
