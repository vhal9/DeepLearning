### Testes para melhorar acur√°cia do modelo para base de cancer de mama

```python
import keras
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

#### Atencao ao diretorio dos arquivos

```python
previsores = pd.read_csv('../../data/base breast cancer/entradas-breast.csv')
classe = pd.read_csv('../../data/base breast cancer/saidas-breast.csv')
```

```python
previsores[:10]
```

### Rede inicial

```python
def criarRede():
    classificador = Sequential()
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform', input_dim = 30))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = 16, activation = 'relu', 
                        kernel_initializer = 'random_uniform'))
    classificador.add(Dropout(0.2))
    classificador.add(Dense(units = 1, activation = 'sigmoid'))
    otimizador = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue = 0.5)
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy',
                      metrics = ['binary_accuracy'])
    return classificador
```

```python
classificador =  KerasClassifier(build_fn = criarRede, epochs = 100, batch_size = 10)
```

```python
resultados = cross_val_score(estimator = classificador,
                             X = previsores, y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
media = resultados.mean()
```

```python
desvio = resultados.std()
```
